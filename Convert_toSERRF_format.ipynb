# This is a script to generate a csv file for SERRF normalization

```python
# import libraries
import pandas as pd
```

```python
# copy XCMS output file path here
input_file = 'XCMS/output_Final-positive.tsv'

# copy output file path here
SERRF_output_file = 'XCMS/SERRF_output_Final-positive.csv'
```

```python
# read in XCMS output file
df_xcms = pd.read_csv(input_file, sep=' ')
df_xcms.head()
```

```python
# modify column name to match SERRF format
df_xcms.rename(columns ={'name': 'Sample'}, inplace=True)
df_xcms.shape
```

# filter features by rt and ioncount
## remove features with rt <= 120s and rt >= 1080s

```python
df_xcms = df_xcms[(df_xcms['rtmed'] >= 120) & (df_xcms['rtmed'] <= 1080)]
df_xcms.shape
```

##  subset xcms output with only required columns for SERRF

```python

# regex to match column names with groups names 
# Sample, QC, WT, KO
for col in df_xcms:
    if col == 'Sample':
        continue
    elif len(col.split('_')) < 3:
        df_xcms.drop(col, axis=1, inplace=True)
print(df_xcms.shape)        
```

```python
df_xcms.head()
```

# remove features with ioncounts <= 10000

```python
# if all the values in a row are <10000, remove the row 
df_xcms = df_xcms[(df_xcms.iloc[:, 1:] >= 10000).any(axis=1)]
df_xcms.shape
```

# Convert to SERRF format

```python
def convertToSerrf(df_xcms):
    df_SERRF = df_xcms.reindex(sorted(df_xcms.columns, key=lambda x: x.split('_')[-1] if x != 'Sample' else x), axis=1)
# set sample column as the first column
    df_SERRF = df_SERRF[['Sample'] + [col for col in df_SERRF.columns if col != 'Sample']]
    idx = [i for i in range(1, len(df_SERRF.index)+1)]
    df_SERRF.index = pd.Index(idx, name='No')
    #print(df_SERRF.head())
# adapt the multiindex format for SERRF
# batch: A, A, A,..
    batch = ['batch'] + ['A'] * (len(df_xcms.columns)-1)
    #print(batch)
# sampleType according to the column name (QC, sample)
    sampleType = []
    for col in df_SERRF.columns:
        if 'Sample' in col:
            sampleType.append('sampleType')
        elif 'QC' in col:
            sampleType.append('qc')
        else:
            sampleType.append('sample')
    print(sampleType)
    # time: injection order (1, 2, 3, ...)
    time = ['time'] + [i for i in range(1, len(df_SERRF.columns))]
    #time = [i for i in range(1, len(df_SERRF.columns))]
    # label: sample name
    label = ['label'] + [col for col in df_SERRF.columns if col != 'Sample']
    #label = [col for col in df_SERRF.columns if col != 'Sample']
    # create multiiindex for the datafile skip the index column
    df_SERRF.columns = pd.MultiIndex.from_arrays([batch, sampleType, time, label])
                                                        
    # add an index column with name 'No' and start from 1
    #df_SERRF.insert(0, 'No', range(1, 1 + len(df_SERRF)))
    return df_SERRF
```

```python
df_SERRF = convertToSerrf(df_xcms)
df_SERRF.head()
```

```python
# save the output file
df_SERRF.to_csv(SERRF_output_file)
```
