# This is a script to generate a csv file for MetaboAnalyst format

```python
# import libraries
import pandas as pd
```

```python
# copy XCMS output file path here

#input_file = 'XCMS/output_Final-positive.tsv'
input_file = 'XCMS/normalized by - SERRF.csv'

# copy output file path here
MetaboAnalyst_output_file = 'XCMS/MetaboAnalyst_output_Final-positive.csv'
```

```python
# read in XCMS output file
df_input = pd.read_csv(input_file, sep=',')
df_input.head()
```

```python
# Rename the Sample 
df_input = df_input.rename(columns={'label': 'Sample'})
```

```python
def get_group(col_name, group):
    for i in group:
        if i == 'QC':
            if i in col_name:
                return 'QC'
        else:
            if i in col_name:
                return i
```

```python
group = ['WT', 'KO', 'QC']
group_row = pd.Series(data=[get_group(col, group) for col in df_input.columns],
                            index=df_input.columns)
df_group =df_input._append(group_row, ignore_index=True)
df_group.tail()
```

```python
def move_row(df):
    # get datafile row
    target_row = df.shape[0] - 1
    # Move target row to first element of list.
    idx = [target_row] + [i for i in range(len(df)) if i != target_row]
    df = df.iloc[idx].reset_index(drop=True)
    df.loc[0,'Sample'] = 'Label'
    return df
```

```python
df_group = move_row(df_group)
df_group.head()
```

```python
df_group.to_csv(MetaboAnalyst_output_file, index=False)
```
