```r
# load xcms output
df_features <- read.csv("XCMS/output_Final-positive.tsv", sep = " ", header = TRUE, check.names = FALSE)
# rename columns mzmed to mz
colnames(df_features)[colnames(df_features) == "mzmed"] <- "mz"
head(df_features)
```

```r
# Load MetaSci
df_metasci <- read.csv("MetaSci_Lib.csv", sep = ",", header = TRUE, check.names = FALSE)
head(df_metasci)
```

```r
library(MetaboAnnotation)
```

```r
parm <- Mass2MzParam(adducts = c("[M+H]+","[M+Na]+"),
                       tolerance = 0, ppm = 20)
```

```r
matched_metasci <- matchValues(df_features, df_metasci, param = parm)
matched_metasci
```

```r
matched_metasci <- matchedData(matched_metasci)[whichQuery(matched_metasci), ]
matched_metasci
```
